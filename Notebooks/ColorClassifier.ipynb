```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('ColorCluster.csv')
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
df.head()
```

```python
# KNN Model to identify 1st cluster
X = df.drop(['image_id', 'Color_3', 'R2', 'G2', 'B2', 'cluster_C1', 'cluster_C2'], axis=1)
y = df.cluster_C1
```

```python
from sklearn.model_selection import train_test_split
## Split data into training and testing sets.
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, test_size = 0.2, random_state = 42)
```

```python
## Import the Classifier.
from sklearn.neighbors import KNeighborsClassifier
## Instantiate the model with 7 neighbors. 
knn_C1 = KNeighborsClassifier(n_neighbors=7)
## Fit the model on the training data.
knn_C1.fit(X_train, y_train)
## See how the model performs on the test data.
knn_C1.score(X_test, y_test)
```

```python
from sklearn.neighbors import NearestCentroid
clf = NearestCentroid()
clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```
